<h1>Functions--------------------------------------------</h1>

```python
## enter the year and month and returns the customer and gross per day of that month
def get_month(year,month): 
    get = str(year)+'-'+str(month)
    return index_date[get]
    
    
```

```python
## enter the year and returns the customer and gross per day of that month
def get_year(year): 
    get = str(year)
    return index_date[get]
    
```

```python
# get the total of any month
def month_total(month,year):
    get = str(year)+'-'+str(month) # converts the input into a string
    month = index_date[get]        # passes it to index_date
    get_month = month.ix[:,'Gross']# pulls just the gross column
    return get_month.sum()         #returns the sum of the column
```

```python
def plot_year(year):
    ### first get months data ###
    empty_list = range(1,13)
    list_of_revenue = []
    for e in empty_list:
       list_of_revenue.append(month_total(e,year))
    ### we now have a list of monthly revenue###
    ###now we plot it ####
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    y = list_of_revenue
    N = len(y)
    ind = range(N)
    ax.bar(ind,y, facecolor = 'B', align='center', ecolor = 'black')
    ax.set_ylabel('REVENUE')
    make_total = str(sum(list_of_revenue)) #This is so we can change input to a string.
    ax.set_title('The Gross Revenue for ' + str(year)+' was $'+ make_total[0:3]+','+make_total[3:],fontstyle='italic') #sums input list and puts it in titles
    ax.set_xticks(ind)
    plt.grid(True)
    group_labels = ['JAN', 'FEB','MARCH','APRIL','MAY','JUNE',
                'JULY', 'AUG','SEPT','OCT','NOV','DEC'
                 ] # labels Y
 
    ax.set_xticklabels(group_labels)
    fig.autofmt_xdate() #see tutorial.
 
    return plt.show()
```

```python
# returns the a horizontal bar graph of clients, sorted with top grossing at top
def top_month(month, year):
    retrieve = get_month(month, year)
    group_retrieve = retrieve.groupby('Customer').aggregate(sum)
    sorts = group_retrieve.sort(columns='Gross', ascending=True)
    sorts.plot(kind='barh')
    xlabel('Revenue')
    month_labels = ['JAN', 'FEB','MARCH','APRIL','MAY','JUNE',
                'JULY', 'AUG','SEPT','OCT','NOV','DEC']
                 
    title('Highest Grossing clients of '+str(month_labels[month-1])+' '+str(year))
    return show()
##TODO:  SPIT OUT TO FILE, OR ON PLOT
```

```python
# returns group room night for anygiven month back to 2011.
def get_grn(month,year):
    group_room['Month']== month
    grn_year = group_room[group_room['Month']== month]
    return grn_year[str(year)]
```

```python
# returns a bar graph of the group room nights back to 2011.
def plot_grn_yearly(year):
    if year < 2011:
        return "Group Room Night Records only go back until 2011"
    else:
        ### first get months data ###
        empty_list = range(1,13)
        list_of_grn = []
        for e in empty_list:
            list_of_grn.append(get_grn(e,year))
        ### we now have a list of monthly revenue###
        ###now we plot it ####
        fig = plt.figure()
        ax = fig.add_subplot(1,1,1)
        y = list_of_grn
        N = len(y)
        ind = range(N)
        ax.bar(ind,y, facecolor = 'B', align='center', ecolor = 'black')
        ax.set_ylabel('Room Nights')
        make_total = str(sum(list_of_grn)) #This is so we can change input to a string.
        ax.set_title('The total Group Room Nights for ' + str(year)+' were '+ make_total[0:2]+','+make_total[2:],fontstyle='italic') #sums input list and puts it in titles
        ax.set_xticks(ind)
        plt.grid(True)
        group_labels = ['JAN', 'FEB','MARCH','APRIL','MAY','JUNE',
                    'JULY', 'AUG','SEPT','OCT','NOV','DEC'
                     ] # labels Y
 
        ax.set_xticklabels(group_labels)
        fig.autofmt_xdate() #see tutorial.
 
        return plt.show()  
```

```python
#Gets the biggest customer of any year given.
def get_biggest_customer(year):
    a = get_year(year)
    combine = a.groupby('Customer').aggregate(sum)
    sort = combine.sort(columns=['Gross'], ascending=False)
    drop = sort['Gross']
    return drop.head(10)
```
